```python
#4.How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Math': [90, 80, 70, 60, 50],
    'Science': [85, 75, 65, 55, 45],
    'English': [88, 82, 76, 70, 65]
}

df = pd.DataFrame(data)
corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#8Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Score': [85, 42, 77, 90, 66]
}

df = pd.DataFrame(data)
threshold = 70

filtered_df = df[df['Score'] > threshold]
print("\nFiltered DataFrame (Score > 70):\n", filtered_df)

```

```python
#12. Generate a bar plot using Plotly.
import plotly.express as px
import pandas as pd

data = {
    'Subject': ['Math', 'Science', 'English', 'History', 'Art'],
    'Marks': [85, 90, 75, 80, 70]
}

df = pd.DataFrame(data)

fig = px.bar(df, x='Subject', y='Marks', title='Student Marks by Subject', color='Subject')

fig.show()



```

```python
#13.Create a 3D scatter plot using Plotly.
import plotly.express as px
import pandas as pd

data = {
    'X': [1, 2, 3, 4, 5],
    'Y': [10, 15, 13, 17, 12],
    'Z': [5, 3, 6, 7, 2],
    'Label': ['A', 'B', 'C', 'D', 'E']
}

df = pd.DataFrame(data)
fig = px.scatter_3d(df, x='X', y='Y', z='Z', color='Label', title='3D Scatter Plot Example')

fig.show()

```
